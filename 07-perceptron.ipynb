```python
# Use perceptrons to model logic gates
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from itertools import product
from sklearn.linear_model import Perceptron


# Create and visualize AND gate
data = [[0, 0], [0, 1], [1, 0], [1, 1]]
labels = [0, 0, 0, 1]

plt.scatter(
  x = [point[0] for point in data], 
  y = [point[1] for point in data],
  c = labels
)
plt.show()
```

```python
# Build perceptron
classifier = Perceptron(max_iter = 40)
classifier.fit(data, labels)
score = classifier.score(data, labels)
print(score)
```

```python
# Visualize the perceptron
# create a list of 100 evenly spaced decimals between 0 and 1
x_values = np.linspace(0, 1, 100)
y_values = np.linspace(0, 1, 100)

# find every possible combination of the x and y values
point_grid = list(product(x_values, y_values))

# the decision_function() method returns the distance those points are from the decision boundary
distances = classifier.decision_function(point_grid)

abs_distances = [abs(d) for d in distances]
distances_matrix = np.reshape(abs_distances, (100, 100))

heatmap = plt.pcolormesh(x_values, y_values, distances_matrix)
plt.colorbar(heatmap)
plt.show()
# The purple line is the decision boundary (where the distances are 0)
```
